# Apache Spark

- Apache Spark, küme üzerinde hızlı ve genel amaç bilgili işleme sistemidir.
- Hadoop'un değil MapReduce yönteminin alternatifi olarak programlanmıştır.
- MapReduce modelinde yer alan disk bazlı çalışma sisteminin yarattığı maliyetlerden dolayı ortaya çıkmıştır.
- Apache Spark, Apache Hadoop'a göre 100 kat daha hızlı çalışmaktadır. (In Memory (RAM - Bellek içi) prensibiyle çalışır.)
- Java, Scala, Python ve R ile uygulama geliştirilebilir.
- Genelleştiricidir: Spark SQL, Spark MLlib, Spark Streaming, GraphX gibi ortamlar aynı uygulama içerisinde kullanılabilir.

# Apache Spark Bileşenleri


- ***Spark Core*** -> Hafıza yönetimi, görevlerin dağıtılması, hata kurtarma, saklama ve dosya sistemlerine erişim sağlar.
- ***RDD's*** -> RDD's MapReduce'un alternatifidir. Apache Spark'ın programlama modeli verinin bellek içi tutularak paralel işlenmesini ifade eder.
- ***Spark SQL*** -> SQL ya da bir dataframe API kullanarak Spark programlarında yapılandırılmış veriyi sorgulama imkanı sunar.
- ***Spark MLLib*** -> Apache Spark'ın ölçeklenebilir makine öğrenmesi kütüphanesidir. Apache Spark'ın bellek içi çalışma mantığından dolayı iteratif işlemler barındıran, makine öğrenmesini gibi işlemlerde bize çözüm sunmaktadır.
- ***Spark Streaming*** -> Akan verinin ölçeklenebilir, yüksek hacimli ve hata toleranslı bir şekilde işlenmesine olanak sağlayan temel Spark API'ının ait bir spark uzantısıdır.
- ***GraphX*** -> Paralel grafik tabanlı hesaplama işlemleri için kullanılan bir kütüphanedir.


# RDD's (Dayanıklı Dağıtık Data Setler)

- RDD'de yapılan işlem, veriyi RAM'e taşıyarak RAM üzerinde dönüşüm işlemlerini, gerekli iteratif işlemleri yapmak ve en son işlemler bittikten sonra veriyi son aşamada tekrar diske yazmaktır.
- 3 aşamadan oluşur:
    - ***RDD's*** -> RDD'sleri oluşturmak. Üretilen bir işlem olduğunda işlemi askıda bekletir (RAM'de tutulur).
    - ***Transformation*** -> Dönüştürme (Mevcut RDD'ler üzerinden dönüştürme işlemi gerçekleştirir.)
    - ***Action*** -> Aksiyon (! Aksiyon basamağı tetiklenmeden uygulama(execution) işlemi gerçekleşmez. Bu sebeble <span style="color:red">"tembel çalışma (lazy evolation)"</span> olarak adlandırılır. Aslında bu tembellik gibi gözükse de hız ve performans arttırıcı bir imkan sağlar.)




- Günlük hayattan RDD's i açıklamak gerekirse:
    - Örneğin bir yemek sofrasına oturduk. Yemek yemeye başladık. Bu esnada annemiz bana mutfaktan çatal getirir misin dedi ve siz de kalktınız çatalı almaya gittiniz. Geri oturdunuz ve masaya koydunuz. Sonra bana kaşık getirir misin dendi ve kalktınız kaşık aldınız. Geri oturdunuz ve masaya koydunuz. Sonra bana tuzluk getirir misin dendi ve kalktınız tuzluğu aldınız. Geri oturdunuz ve masaya koydunuz. İşte bu hadoop'un çalışma mantığıdır. Spark ise bunu istemez. 
    - Spark'ta ise kalktınız çatalı aldınız elinizde (distrubuted memory - RAM) bekliyorsunuz bekliyorsunuz. Kaşık istendiğinde kaşığı da alıyorsunuz elinizde (distrubuted memory - RAM) bekliyorsunuz. Tuzluk istendiğinde tuzluğu da alıyorsunuz elinizde (distrubuted memory - RAM) bekliyorsunuz bekliyorsunuz. En son bir şey istenmediğini ifade eden bir aksiyon komutu geldiğinde askıda bekletilen bu iteratif işlemleri sonlandırıp sofraya oturuyorsunuz.

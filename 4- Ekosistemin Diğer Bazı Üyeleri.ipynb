- Burada bu büyük veri ekosistemini oluşturan yazılımların hangi amaçlar çerçevesinde konumlandığını değerlendireceğiz.
    - ***ZooKeeper*** -> Apache Hadoop kümelerinin kurulumu, yönetimi ve gözetimi için oluşturulmuş web tabanlı araçlardır. Kümelerin durumunu, performanslarını takip edecek bir gösterge panelini kullanır.
    - ***Hıve & Drill*** -> Arka planda sql komutlarını MapReduce'e çevirerek büyük veride sql sorgusu atabilmemizi sağlar.
    - ***Kafka & Storm*** -> Büyük veri akışını düşük bir gecikmeyle gerçek zamanlı sağlayabilen araçtır. Linkedin bünyesinde geliştirilmiştir.
    - ***Mahoot & Spark MLLib*** -> Makine öğrenmesi görevleri için geliştirilmiştir. Ama Spark'taki son gelişme ve performans artışlarından sonra Mahoot geliştirmeleri durdurulmuştur.
    - ***PIG*** -> SQL türevi daha scripting kısmına bakacak yönüyle bazı görevleri yerine getirmek için kullanılır.
    - ***SOLR & LUCENE*** -> Arama ve indexleme işlemleri için kullanılır.
    - ***HBase*** -> NoSQL veri tabanı yönetim sistemidir. Büyük boyutlu verilerde gerçek zamanlı okuma yazma işlemi yapmamızı sağlar. Sütun bazlıdır.
    - ***Oozie*** -> İş planlaması için kullanılır.
    - <span style="color:red"><b>PySpark</b></span> -> Python'da Spark ile çalışabilmek için kullanacağımız kütüphanedir.

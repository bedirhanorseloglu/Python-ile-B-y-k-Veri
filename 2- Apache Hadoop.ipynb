# Apache Hadoop

- Apache Hadoop; açık kaynak kodlu, güvenlir, ölçeklenebilir paralel hesaplama yazılım projesidir.
- Hadoop ismi aslında bu projeyi tasarlayın kişinin çocuğunun oyuncak filinin adından gelmektedir.
- Apache Hadoop büyük veri dünyasının temelini oluşturmaktadır.
- Bir bilgisayar kümesinin belirli bir işi yapmak için tek bir bilgisayar gibi birlikte hareket etmesini sağlamaktadır.

# Apache Hadoop Bileşenleri

- ***Hadoop Common*** -> Büyük veri teknolojileri, diğer ifadesi ile tüm hadoop modüllerini destekleyen ortak gereksinimlerdir.
- ***Hadoop Dağıtık Dosya Sistemi (HDFS)*** -> Yerel dosya sistemleri olan FAT43 ve NFTS gibi dosya sistemidir. Bu sistemlerden farkı büyük boyutlardaki veriyi dağıtık şekilde depolamaya ve kontrol etmeye imkan sağlamaktadır.
- ***Hadoop YARN*** -> Kaynak yönetimi ve iş planlaması için kullanılan Apache Hadoop'un daha etkin bir şekilde kullanılmasına olanak sağlayan bileşendir.
- ***Hadoop MapReduce*** -> Büyük veri dünyasının fonksiyonel anlamda temelini oluşturan bileşendir. Aynı ağdaki dağıtık bilgisayar kümeleri üzerinde büyük veri analizi yapılabilmesi için geliştirilmiş bir programlama modelidir.

# MapReduce Nedir

- Görevimiz inputta verilen hayvan isimlerinden hangi hayvandan kaç tane var bunu bulmak. 
- Buna yönelik olarak önce ayırma(split) işlemi yaparak her satırı böldük.
- Sonrasında map aşamasında her bir splitteki köpek, kedi ve fareler kendi içlerinde sayılmış ve yeni veri parçacıkları oluşturmuştur. (Mapping aşamasının görevi HDFS üzerindeki girdi verilerini işlemektir.)
-  Shuffling basamağında ise map basamağında yapılan işlemler düzenleştirilecektir. Aslında Shuffling vve Reducing bir arada hareket eder. Burada mapping aşamasında tanımlanmış olan köpekler, kediler ve bunların sayıları shuffle ve reduce basamağında map aşamasının her basamağında bulunan sonuçlar bir araya getirilerek değerlendirme yapılıyor. (Shuffle ve Reduce aşamalarının görevi, Map aşamasından gelen veriyi işlemek ve indirgemektir.)
- Son basamakta ise reduce'taki işlemler bir araya getirilir.

# Apache Hadoop Çözüm Olarak Yeterli Mi?


- Disk tabanlı çalışan bir modeldir. Yani her MapReduce görevinde (her bir iterasyonda) diskten okuma ve yazma işlemi yapılır. Bu yüzden iteratif işlemler zaman alır ve kaynakları meşgul eder.
- <span style="color:red">Bu sebeple disk bazlı çalışma modeli yerine RAM bazlı çalışma modeli olarak Hadoop'a alternatif olarak değil, MapReduce Disk Bazlı Çalışma Modeline alternatif olarak Apache Spark adını verdiğimiz bir çalışma prensibi oluşturulmuştur.</span>



- Günlük hayattan MapReduce'u açıklamak gerekirse:
    - Örneğin bir yemek sofrasına oturduk. Yemek yemeye başladık. Bu esnada annemiz bana mutfaktan çatal getirir misin dedi ve siz de kalktınız çatalı almaya gittiniz. Geri oturdunuz ve masaya (on disk) koydunuz. Sonra bana kaşık getirir misin dendi ve kalktınız kaşık aldınız. Geri oturdunuz ve masaya koydunuz. Sonra bana tuzluk getirir misin dendi ve kalktınız tuzluğu aldınız. Geri oturdunuz ve masaya koydunuz. İşte bu hadoop'un çalışma mantığıdır. Spark'taki RDD's ise bunu istemez.
